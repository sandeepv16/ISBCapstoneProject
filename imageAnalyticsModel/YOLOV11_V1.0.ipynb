```python
#Predict using best model
model = YOLO(r"C:\Users\ADMIN\SV_Practice_06102024\CAPSTONE_BASELINE\YOLOV11\runs\detect\train7\weights\best.pt")
results=model.predict(source=r"C:\Capstone Project\AiSPRY Datasets\pipe_data\frames\CSH.jpg",save=True,conf=0.15,max_det=1000)
```

```python
#Print count of pipes per class
c = 0
h = 0
s1  = 0
s2 = 0
for box in results[0].boxes:
    class_id = int(box.cls)
    classname = model.names[class_id]
    if classname == 'c':
        c = c+1
    if classname == 'h':
        h = h+1
    if classname == 's1':
        s1 = s1+1
    if classname == 's2':
        s2 = s2+1
print('Circular : ' + str(c) + ' Hex : ' + str(h) + ' Square 1 : ' + str(s1) + ' Square 2: ' + str(s2))
```
